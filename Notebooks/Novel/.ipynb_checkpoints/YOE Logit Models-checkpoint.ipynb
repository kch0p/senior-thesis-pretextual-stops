### MODEL

#### SDPD

```python
SDPD = pd.read_csv('ALL-SDPD-AB953-DUSK.csv')
group = 'hispanic/latino/a'

SDPD_VOD = SDPD[(SDPD['itp']==1) & 
                ((SDPD['mins_before_dusk']>30) | (SDPD['mins_before_dusk']<0)) &
                ((SDPD['race']=='white')|(SDPD['race']==group))]

#https://www.geeksforgeeks.org/ml-one-hot-encoding-of-datasets-in-python/
vod_SD_dusk = pd.get_dummies(SDPD_VOD, columns = ['race'], drop_first=False, prefix='')\
                       [['date',
                         'mins_stop_time',
                         'location',
                         'Experience',
                         'dark',
                         f'_{group}']].reset_index(drop=True)\
                                                    .rename(columns={f'_{group}':'minority'})


# vod_SF_dusk['Experience'] = vod_SF_dusk['Experience'].astype(int)
vod_SD_dusk['mins_stop_time'] = vod_SD_dusk['mins_stop_time'].astype(int)
vod_SD_dusk['location'] = vod_SD_dusk['location'].astype(str)
```

```python
sd_loc_table = vod_SD_dusk.groupby('location',as_index=False).agg('count')\
                                                             .rename(columns={'date':'count'})\
                                                             .iloc[:,:2]
sd_loc_table.describe()
sd_loc_table['freq'] = ''
sd_loc_table['count'].describe()
sd_loc_table.loc[sd_loc_table['count']<75,'freq']='low_freq'
sd_loc_table.loc[(sd_loc_table['count']>=75)&(sd_loc_table['count']<=275),'freq']='average'
sd_loc_table.loc[(sd_loc_table['count']>275),'freq']='high_freq'
sd_loc_table = sd_loc_table.sort_values(by='count',ascending=False).reset_index(drop=True)
sd_loc_table.head()

vod_SD_dusk = vod_SD_dusk.merge(sd_loc_table[['location','freq']],on='location')


def custom_round(x, base=5):
    return int(base * round(float(x)/base))

vod_SD_dusk['mins_stop_time'] = [custom_round(i,base=5) for i in vod_SD_dusk['mins_stop_time']]

vod_SD_dusk.head()
```

```python
# Load your data
df = vod_SD_dusk  # The DataFrame containing the independent and dependent variables

# Create a natural spline with 6 degrees of freedom using the time variable and the dependent variable
print('debug, pre-spline')
spline = UnivariateSpline(df.index.values, df['mins_stop_time'], k=5, s=0.1)
print('debug, post-spline')



# Add the natural spline as a predictor to the DataFrame
df['time_spline_df6'] = spline(df.index)

# Use the `glm` function to fit a logistic regression model with the other independent variables and the natural spline
model = glm('minority ~ C(dark) + C(freq) + Experience + time_spline_df6', data=df, family=sm.families.Binomial())

# Fit the model and display the summary
log_reg = model.fit()
results = log_reg.summary()
OR = pd.DataFrame(
    {
        "OR": log_reg.params,
        "Lower CI": log_reg.conf_int()[0],
        "Upper CI": log_reg.conf_int()[1],
    },
)
OR = np.exp(OR)

print(f'{OR}\n\n\n{results}\n\n\n\n\n\n\n\n\n')
df
```

```python
import pandas as pd
import statsmodels.api as sm
from scipy.interpolate import InterpolatedUnivariateSpline

# Load the data
data = vod_SD_dusk

# Add a constant term to the data
data = sm.add_constant(data)

# Create dummy variables for any categorical variables
data = pd.get_dummies(data, columns=['location', 'dark'])
data = data.sort_values('mins_stop_time')
data = data.drop_duplicates('mins_stop_time')

# Create the design matrix
X = data.values

# Create an InterpolatedUnivariateSpline transformation with 6 degrees of freedom
transformer = InterpolatedUnivariateSpline(data['mins_stop_time'], data['mins_stop_time'], k=3, ext='const')

# Transform the variable and add it to the design matrix
# X = np.column_stack((X, transformer.fit_transform(data['mins_stop_time'])))
X = np.column_stack((X, transformer.transform(data['mins_stop_time'])))
mod = sm.MNLogit(data['minority'], X)


res = mod.fit()
results = res.summary()
OR = pd.DataFrame(
    {
        "OR": res.params,
        "Lower CI": res.conf_int()[0],
        "Upper CI": res.conf_int()[1],
    },
)
OR = np.exp(OR)

print(f'{OR}\n\n\n{results}\n\n\n\n\n\n\n\n\n')
```

```python
SFPD = pd.read_csv('ALL-SFPD-AB953-DUSK-LOC.csv')

group = 'black/african american'
SFPD_VOD = SFPD[(SFPD['itp']==1)&
                ((SFPD['mins_before_dusk']>30) | (SFPD['mins_before_dusk']<0)) &
                ((SFPD['race']==f'{group}')|(SFPD['race']=='white'))]


#https://www.geeksforgeeks.org/ml-one-hot-encoding-of-datasets-in-python/
vod_SF_dusk = pd.get_dummies(SFPD_VOD, columns = ['race'], drop_first=False, prefix='')\
                       [['date',
                         'mins_stop_time',
                         'zip',
                         'Experience',
                         'dark',
                         f'_{group}']].reset_index(drop=True)\
                                                    .rename(columns={f'_{group}':'black',
                                                                      'zip':'location'})


# vod_SF_dusk['Experience'] = vod_SF_dusk['Experience'].astype(int)
vod_SF_dusk['mins_stop_time'] = vod_SF_dusk['mins_stop_time'].astype(int)
vod_SF_dusk['location'] = vod_SF_dusk['location'].astype(str)

pop_zips = pd.read_csv('sf_pops_zips.csv').iloc[1:,:].rename(columns={'zip':'location'})
vod_SF_dusk = vod_SF_dusk.merge(pop_zips,on='location').sort_values(by='date').reset_index(drop=True)
vod_SF_dusk['pop'] = vod_SF_dusk['pop'].astype(int)

vod_SF_dusk.head()
```

```python
pd.options.display.float_format = '{:.3f}'.format
formula = f"black ~ C(dark,Treatment(reference=0)) \
                  + mins_stop_time \
                  + C(freq,Treatment(reference=0)) \
                  + C(pop_cat,Treatment(reference=0)) \
                  + Experience"
log_reg = smf.logit(formula, data = vod_SF_dusk).fit()
results = log_reg.summary()
OR = pd.DataFrame(
    {
        "OR": log_reg.params,
        "Lower CI": log_reg.conf_int()[0],
        "Upper CI": log_reg.conf_int()[1],
    },
)
OR = np.exp(OR)

print(f'{OR}\n\n\n{results}\n\n\n\n\n\n\n\n\n')
```
